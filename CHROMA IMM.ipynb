```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_path = '/Users/DELL/OneDrive/Desktop/DSE/UrbanSound8K/audio/'
```

```python
#godbark
dogbark='/Users/DELL/OneDrive/Desktop/DSE/UrbanSound8K/audio/fold5/100032-3-0-0.wav'
```

```python
data, sample_rate=librosa.load(dogbark)
```

```python
chroma_stft=librosa.feature.chroma_stft(y=data, sr=sample_rate,n_chroma=40)
```

```python
chroma_stft.shape
```

```python
librosa.display.specshow(chroma_stft,
                        x_axis = "time",
                        sr=sample_rate)
plt.title("Chromagram of a dog bark sound")
plt.colorbar(format='%+2f')
plt.show()
plt.savefig("/Users/DELL/OneDrive/Desktop/DSE\Machine Learning/croma1.jpg")
```
